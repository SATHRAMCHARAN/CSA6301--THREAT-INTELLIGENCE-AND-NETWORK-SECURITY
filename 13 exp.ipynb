```python
# File Integrity Checker using SHA-256 Hash Comparison

import hashlib

def calculate_hash(file_name):
    sha256 = hashlib.sha256()

    with open(file_name, "rb") as file:
        while chunk := file.read(4096):
            sha256.update(chunk)

    return sha256.hexdigest()


file_name = "sample.txt"

# Create sample file
with open(file_name, "w") as file:
    file.write("This is the original file content.")

# Generate known-good baseline hash
baseline_hash = calculate_hash(file_name)

print("Known-Good Baseline Hash:")
print(baseline_hash)

# Modify file for testing
choice = input("\nModify the file? (yes/no): ")

if choice.lower() == "yes":
    with open(file_name, "w") as file:
        file.write("This file has been modified.")

# Calculate current hash
current_hash = calculate_hash(file_name)

print("\nCurrent File Hash:")
print(current_hash)

# Compare hashes
if baseline_hash == current_hash:
    print("\nFile Integrity Verified: File has not been modified.")
else:
    print("\nALERT: File has been modified!")
```
