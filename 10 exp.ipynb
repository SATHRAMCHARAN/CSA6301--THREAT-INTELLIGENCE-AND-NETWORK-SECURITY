```python


failed_login_threshold = 3

login_logs = [
    "Failed login from 192.168.1.10",
    "Failed login from 192.168.1.10",
    "Failed login from 192.168.1.10",
    "Successful login from 192.168.1.20",
    "Failed login from 10.0.0.5",
    "Failed login from 10.0.0.5"
]

failed_attempts = {}

for log in login_logs:
    if "Failed login" in log:
        ip = log.split()[-1]

        if ip in failed_attempts:
            failed_attempts[ip] += 1
        else:
            failed_attempts[ip] = 1


print("IDS Scan Results:\n")

for ip, count in failed_attempts.items():
    if count >= failed_login_threshold:
        print("ALERT: Possible Brute Force Attack Detected!")
        print("IP Address:", ip)
        print("Failed Login Attempts:", count)
    else:
        print("Normal Activity - IP:", ip, "Failed Attempts:", count)
```
