```python
import re

def check_phishing_url(url):
    suspicious_keywords = [
        "login", "verify", "update", "secure",
        "account", "bank", "password", "confirm"
    ]

    score = 0

    if not url.startswith("https://"):
        score += 1

    for keyword in suspicious_keywords:
        if keyword in url.lower():
            score += 1

    if len(url) > 75:
        score += 1

    if "@" in url or "-" in url:
        score += 1

    if re.search(r"\d+\.\d+\.\d+\.\d+", url):
        score += 1

    print("URL:", url)

    if score >= 3:
        print("Result: Suspicious URL - Possible Phishing Link")
    else:
        print("Result: URL appears safe")


url = input("Enter URL: ")
check_phishing_url(url)
```
