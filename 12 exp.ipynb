```python
# Automatic Incident Report Generator

from datetime import datetime

security_events = [
    {
        "event_id": 1,
        "type": "Brute Force Attack",
        "source_ip": "192.168.1.10",
        "severity": "High",
        "description": "Multiple failed login attempts detected"
    },
    {
        "event_id": 2,
        "type": "Malware Detection",
        "source_ip": "10.0.0.5",
        "severity": "Critical",
        "description": "Malicious file execution detected"
    }
]

report_file = "incident_report.txt"

with open(report_file, "w") as report:
    report.write("SECURITY INCIDENT REPORT\n")
    report.write("=======================\n")
    report.write("Generated Time: " + str(datetime.now()) + "\n\n")

    for event in security_events:
        report.write("Incident ID: " + str(event["event_id"]) + "\n")
        report.write("Event Type: " + event["type"] + "\n")
        report.write("Source IP: " + event["source_ip"] + "\n")
        report.write("Severity: " + event["severity"] + "\n")
        report.write("Description: " + event["description"] + "\n")
        report.write("-----------------------\n")

print("Incident report generated successfully.")

with open(report_file, "r") as report:
    print("\n" + report.read())
```
