```python
import socket

def resolve_domains(domains):
    print("Domain Name to IP Address:\n")

    for domain in domains:
        try:
            ip = socket.gethostbyname(domain)
            print(domain, "->", ip)
        except socket.gaierror:
            print(domain, "-> Unable to resolve")


# List of domains
domains = [
    "google.com",
    "amazon.com",
    "github.com",
    "microsoft.com"
]

# Resolve IP addresses
resolve_domains(domains)
```
