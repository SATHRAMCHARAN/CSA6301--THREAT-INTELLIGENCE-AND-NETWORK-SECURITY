```python


malicious_ips = [
    "192.168.1.100",
    "10.0.0.50",
    "185.220.101.5"
]

traffic_logs = [
    "Connection from 192.168.1.100 to server",
    "User login from 172.16.0.10",
    "Blocked request from 185.220.101.5",
    "Normal traffic from 8.8.8.8",
    "Connection attempt from 10.0.0.50"
]

print("Scanning Traffic Logs...\n")

found = False

for log in traffic_logs:
    for ip in malicious_ips:
        if ip in log:
            print("ALERT: Malicious IP detected!")
            print("IP Address:", ip)
            print("Log Entry:", log)
            print()
            found = True

if not found:
    print("No malicious IP addresses detected.")
```
