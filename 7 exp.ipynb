```python


from collections import defaultdict

# Create sample login log file
log_file = "login_log.txt"

with open(log_file, "w") as file:
    file.write("""Failed login attempt from 192.168.1.10
Successful login from 10.0.0.5
Failed login attempt from 192.168.1.10
Failed login attempt from 172.16.0.20
Failed login attempt from 192.168.1.10
Failed login attempt from 172.16.0.20
""")


failed_attempts = defaultdict(int)

# Read and analyze log file
with open(log_file, "r") as file:
    for line in file:
        if "Failed login" in line:
            ip = line.split()[-1]
            failed_attempts[ip] += 1


print("Failed Login Attempts per IP Address:\n")

for ip, count in failed_attempts.items():
    print(ip, ":", count, "failed attempts")
```
