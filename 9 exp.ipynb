```python


firewall_rules = [
    {"ip": "192.168.1.10", "port": 80, "action": "ALLOW"},
    {"ip": "10.0.0.5", "port": 22, "action": "BLOCK"},
    {"ip": "192.168.1.20", "port": 443, "action": "ALLOW"}
]

def firewall_check(packet):
    for rule in firewall_rules:
        if packet["ip"] == rule["ip"] and packet["port"] == rule["port"]:
            return rule["action"]

    return "BLOCK"

packet = {
    "ip": input("Enter source IP address: "),
    "port": int(input("Enter port number: "))
}

decision = firewall_check(packet)

print("\nPacket Details:")
print("Source IP:", packet["ip"])
print("Port:", packet["port"])
print("Firewall Decision:", decision)
```
